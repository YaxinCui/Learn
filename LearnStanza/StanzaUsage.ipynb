![avatar](https://stanfordnlp.github.io/stanza/assets/images/pipeline.png)

````python
import stanza

stanzaNLP = stanza.Pipeline('es', processors="lemma, pos,tokenize,depparse")
text='En 9R nos trataron genial, Julio era atento, cordial y divertido, nos trató de maravilla al igual q los demás.'

document = stanzaNLP(text)
````

````python
document.entities
````
