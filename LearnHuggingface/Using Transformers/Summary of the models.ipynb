## Summary of the models

将讲述对于不同的任务，会基于什么模型来微调。
- Decoders or autoregressive models
  - Original GPT：第一个基于transformer的预训练结构，使用Book Corpus数据集
  - GPT-2：更大版本的GPT，使用webText训练
  - CTRL：可以结构化控制生成输出的GPT
  - Transformer-XL：log(n)级attention、动态位置编码
  - Reformer：
    - 使用axial position encoding
    - 使用lsh attention替代点积attention
    - 使用可逆transformer，避免存储中间结果
    - 按块进行前向运算，而不是整批运算
  - XLNet：使用permutes mask，使用transformer-xl
