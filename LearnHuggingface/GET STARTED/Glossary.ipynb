### 基本术语
- 自编码模型：MLM
- 自回归模型：CLM
- CLM：因果语言模型，通过前文预测后文
- MLM：mask任务
- multimodal：多模态任务，输入有多种模态
- NLG：自然语言生成任务
- NLP：自然语言处理
- NLU：自然语言理解，理解整句话，而不是单个词
- 预训练模型
- RNN：
- 自注意力：
- seq2seq：输入文本进入模型，输出文本
- token：句子的一部分，可以是一个词，也可以是子单词
- transformer：基于自注意力机制的深度学习框架

### Model Inputs | 模型输入
Input IDs

attention_mask

Token Type IDs

Position IDs: 用来识别每个token的位置，可以有绝对位置、也有相对位置

Labels：模型的可选参数，对于不同任务，label也不同

Decoder input IDs：对于encoder-decoder结构，训练时会创建id。

Feed Forward Chunking：用来加速模型训练，在模型消耗内存、时间复杂度中做权衡。
