### 目前支持的模型

在对自然语言表示进行预训练时，增加模型大小通常会提高下游任务的性能。然而，由于GPU/TPU内存限制和更长的训练时间，在某些时候进一步增加模型变得更加困难。为了解决这些问题，我们提出了两种参数缩减技术，以降低内存消耗并提高训练速度。综合的经验证据表明，我们提出的方法可以得到比原始方法更好的模型。我们还使用了一个自我监督的损失模型，重点是句子间连贯性的建模，并表明它始终有助于多句子输入的下游任务。因此，我们的最佳模型在GLUE、RACE和\ squad基准上建立了新的最先进的结果，同时与BERT large相比参数更少。代码和预先训练的模型可以在这个https URL上获得。
